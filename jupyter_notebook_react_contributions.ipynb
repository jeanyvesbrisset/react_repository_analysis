# Analyse et Visualisation des contributions du repo React en Json

```python
import pandas as pd
import json
from matplotlib import pyplot as plt
    
data = pd.read_json("react_contrib_2019-12-30.json")

plt.plot(data[2],data[1])
plt.plot(loglog=True)
plt.title('Proportions des contributeurs par rapport aux autres contributeurs ayant realisés le même nombre de contributions')
plt.xlabel('Pourcentage')
plt.ylabel('nombre de contributions')
plt.show()
```
